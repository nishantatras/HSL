```python
import cv2

# Initialize the webcam
cap = cv2.VideoCapture(0)  # '0' is the default ID for the primary webcam

# Check if the webcam is opened correctly
if not cap.isOpened():
    print("Error: Could not open webcam.")
    exit()
count=1
while (count!=101):
    # Capture frame-by-frame
    ret, frame = cap.read()

    # Check if the frame was captured correctly
    if not ret:
        print("Error: Could not read frame.")
        break

    # Display the frame
    cv2.imshow('Press Space to Capture, Q to Quit', frame)

    # Wait for a key press
    key = cv2.waitKey(1) & 0xFF

    if key == ord(' '):  # Space key to capture the image
        cv2.imwrite(f'{count}.jpg', frame)
        print(f"Image captured and saved as {count}.jpg")
        count+=1

    elif key == ord('q'):  # 'q' key to exit the loop
        print("Exiting...")
        break

# Release the webcam and close the window
cap.release()
cv2.destroyAllWindows()
```

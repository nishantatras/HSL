```python
import matplotlib.pyplot as plt

# Your function to load the image
from PIL import Image
import numpy as np

def simple_vision(img_path):
    img = Image.open(img_path)
    img = img.convert("RGB")
    img_array = np.array(img)
    return img_array

# Load the image using your function
filename = r"C:\\Users\\mehro\\Desktop\\.jpeg"
Image_Sign = simple_vision(filename)

# Create a figure using matplotlib
plt.figure(figsize=(10, 10))

# Display the image
plt.imshow(Image_Sign)
plt.xlabel("")
plt.show()
```
